```python
## Accessing List Elements

fruits=["apple", "banana", "cherry","kiwi"]
fruits[3]
```

```python
fruits[2]
```

```python
fruits[1]
```

```python
print(fruits[2])  # This will raise an IndexError since there is no index 4 in the list
```

```python
print(fruits[1:])
print(fruits[:2])
print(fruits[-1])
print(fruits[::1])
print(fruits[1:3])
```

```python
##modifying list
fruits[3]="orange"  # This will raise an IndexError since there is no index 4 in the list
print(fruits)
```

```python
fruits.append("grapes")
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'orange', 'grapes']
```

```python
fruits.insert(2,"orange")
```

```python
print(fruits[3])  # Output: kiwi
print(fruits)
```

```python
fruits.remove("grapes")
print(fruits)
```

```python
popped_fruit = fruits.pop(2)  # Removes and returns the item at index 2
print(popped_fruit)  # Output: orange   
print(fruits)  # Output: ['apple', 'banana', 'kiwi']
```

```python
fruits.reverse()
print(fruits)  # Output: ['kiwi', 'banana', 'apple']
```

```python
## slicing list
numbers = [1, 2, 3, 4, 5]
print(numbers[1:3])  # Output: [2, 3]
```

```python
list=[]
for x in range(10):
    list.append(x**2)
    print(list)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
[x**2 for x in range(10)]  # List comprehension to create a list of squares
```

```python
### lit comprehension
 # basic syntax  [expression for item in iterable if condition ]
# with conditional logic


squares = [x**2 for x in range(10) ]
print(squares)
```

```python
list =[]
for i in range(10):
    if i%2==0:
        list.append(i)
print(list)        
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
pair=[(i,j) for i in list1 for j in list2]
print(pair)
```
