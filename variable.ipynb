```python
##loop control statememnt

#break
#break statement exits the loop immediately

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
##continue
#continue statement skips the current iteration and continues with the next one
for i in range(10):
    if i %2== 0:
        continue
    print(i)    
```

```python
##pass
#pass statement does nothing and is used as a placeholder
#passs statement is a null statement that is used when a statement is syntactically required but you don't want to execute any code
for i in range(10):
    if i  == 3:
        pass  # Placeholder for future code
    print(i)
```

```python
## nested loops
## a loop inside a loop
for i in range(3):  # Outer loop
    for j in range(2):  # Inner loop
        print(f"i: {i}, j: {j}")
```

```python
## Examples calculate the sum of first N natural number using a while and for loop
n=10
sum=0
count=1
while count <= n:
    sum += count
    count += 1
print ("sum of first 10 natural nummber:", sum)    
```
