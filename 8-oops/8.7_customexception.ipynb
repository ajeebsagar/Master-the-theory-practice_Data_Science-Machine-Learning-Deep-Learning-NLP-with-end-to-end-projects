Custom exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass

class CustomError(Error):
    pass
```

```python
class dobException(Exception):
    pass

try:
    year = int(input("Enter your year of birth: "))
    current_year = 2023  # Or use datetime.datetime.now().year for dynamic year
    age = current_year - year

    if 20 <= age <= 30:
        print("The age is valid, so you can register for the exam.")
    else:
        raise dobException

except dobException:
    print("Sorry, the age is not valid, so you cannot register for the exam.")
except ValueError:
    print("Invalid input. Please enter a valid year.")
```
