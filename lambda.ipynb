lambda function

```python
def additon(a, b):
    """Return the sum of two numbers."""
    return a + b

# Call the function with correct spelling
print(additon(2, 3))

# Define a lambda function for addition
addition = lambda a, b: a + b
print(type(addition))
print(addition(2, 3))
```

```python
addition=lambda x,y,z:x+y+z
addition(1,2,3)
print(addition(1,2,3))
```

```python
## map()  it applied a function to all items in a list

numbers=[1,2,3,4,5]
def square(x):
    """Return the square of a number."""
    return x **2
square(2)

```

```python
def squaree(x):
    return x*x
print(list(map(squaree, numbers)))
```
