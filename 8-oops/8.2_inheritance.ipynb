Inheritance in python

Inheritance is a fundamental concept in Object-Oriented Programming(oop) that allow a class to inherit attributes and methods from another class. This lesson covers single inheritance and multiiple inheritance demona=strating how to crate and them in python.

```python
### inheritance
## parent class

class Car:
    def __init__(self, windows, doors, color):
        self.windows = windows
        self.doors = doors
        self.color = color


    def drive(self):    
        print("The car is driving.")        
```

```python
car1=Car(4, 4, "red")
car1.drive()    
print(f"Car has {car1.windows} windows, {car1.doors} doors, and is {car1.color}.")
```

```python
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, color, autopilot):
        super().__init__(windows, doors, color)
        self.autopilot = autopilot

    def drive(self):
        print("The Tesla is driving with autopilot enabled.")  

Tesla1 = Tesla(4, 4, "blue", True)
Tesla1.drive()
print(f"Tesla has {Tesla1.windows} windows, {Tesla1.doors} doors, is {Tesla1.color}, and autopilot is {Tesla1.autopilot}.")        
```

```python
## multiple inheritance
## when a class inherits from more than one base class.

class Animals:
    def __init__ (self, name):
        self.name = name

    def speak(self):
        print("subclass must implement this method")


 ## base class 2

class Pets:
    def __init__(self, owner):
        self.owner = owner


        ##derived class
class Dog(Animals, Pets):
    def __init__(self, name, owner):
        Animals.__init__(self, name)
        Pets.__init__(self, owner)

    def speak(self):    
        return f"{self.name} says Woof! and is owned by {self.owner}"

    ## create an object
dog1 = Dog("Buddy", "Ajeeb sagar")
print(dog1.speak())    
print(f"{dog1.name} is owned by {dog1.owner}.")      
```
