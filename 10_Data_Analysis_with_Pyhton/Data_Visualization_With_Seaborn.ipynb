Data Visualization With Seaborn

Seaborn is a python visualization library based on Mathplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. in this leson, we will cover the basics of seaborn, including creating various types of plots and customizong them.

```python
import seaborn as sns


###basic plotting with seaborn

tips=sns.load_dataset('tips')
tips
```

```python
## create a scatter plot

import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
## line plot

sns.lineplot(x='size', y='total_bill', data=tips, hue='day')
plt.title('Line Plot of Total Bill vs Size')
plt.xlabel('Size of the Group')
```

```python
## bar plot
sns.barplot(x='day', y='total_bill', data=tips, hue='smoker')
plt.title('Bar Plot of Total Bill by Day and Smoker Status')
```

```python
## categorical plot
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='smoker')
plt.title('Box Plot of Total Bill by Day and Smoker Status')
```

```python
## Box Plot
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
plt.title('Box Plot of Total Bill by Day and Smoker Status')    
plt.show()
```

```python
## violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
```

```python
## histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill') 
```

```python
## kde plot
sns.kdeplot(tips['total_bill'], shade=True)
plt.title('KDE Plot of Total Bill')
```

```python
## pairplot
sns.pairplot(tips, hue='smoker')
```

```python
sns.pairplot(tips)
```

```python
## heat map
corr=tips[['total_bill', 'tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## plot total sales by product

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum, ci=None)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')       
plt.ylabel('Total Sales')
plt.show()
```

```python
## plot total sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum, ci=None)
plt.title('Total Sales by Region')
plt.xlabel('Region')       
plt.ylabel('Total Sales')
plt.show()
```
