Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexiy and effort.

```python
from abc import ABC, abstractmethod
### Abstract base class

class Vehicle(ABC):
    def drive(self):
        print("the vehicle is used for driving")

    @abstractmethod
    def fuel_type(self):
        pass   
class Car(Vehicle): 
    def fuel_type(self):
        print("the car uses petrol")

def operate_vehicle(vehicle: Vehicle):
    vehicle.drive()
    vehicle.fuel_type()

car = Car()
operate_vehicle(car)        
```

```python
###  basic magic methods
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
person = Person("Ajeeb sagar", 30) 
print(person )  
```

```python
###  basic magic methods
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"    
person = Person("Ajeeb sagar", 30) 
print(person )  
print(repr(person))
print(str(person))
```
