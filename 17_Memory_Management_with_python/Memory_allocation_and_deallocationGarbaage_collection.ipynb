### Refrence Counting

reference counting is the primary method python uses to manage memory. each maintains a count of references pointing to it. when the reference count ddrops to zero, the memory occupied by the object is deallocated.

```python
import sys

a=[]
print(sys.getrefcount(a))
```

```python
b=a
print(sys.getrefcount(b))
```

```python
del b
print(sys.getrefcount(a))
```

### garbage collection in python
pyhton includes a cyclic garbage collector to handle reference cycles. reference cycles occur when two or more objects reference each other, preventing their reference counts from reaching zero. The garbage collector periodically identifies and collects these cycles, freeing up memory that is no longer accessible.

```python
import gc
gc.enable
```

```python
gc.disable()
```

```python
gc.collect()
```

```python
### get garbage collection stats

print(gc.get_stats())
```

```python
## get unreachable objects
print(gc.garbage)
```

### memory managment best practices

1. use local variables: local variables have a shorter lifespan and are freed sooner than global variables.
2. avoid circular references: circular references can lead to memory leaks if not properly managed.
3. Use Generators: Generators produces items oe at a time and only keepone item in memory at a time, making them memory efficient.
4. explicity delete objects: use the del statement to delete variables and objects explicity.
5. profiles memory usage: use meomory profilling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.


```python
### handling circular references
import gc
class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created.")

    def __del__(self):
        print(f"Object {self.name} deleted.")    
        

 ####   create circular reference

obj1=MyObject("obj1")
obj2=MyObject("obj2")
obj1.ref=obj2
obj2.ref=obj1      

del obj1
del obj2


### manually trigger the grabge collection
gc.collect()
```

```python
## generator for memory efficient data processing
def generate_numbers(n):
    for i in range(n):
        yield i

## using the generator
for number in generate_numbers(100000):
    print(number)        
    if number>=10:
        break
```

```python
## profilling memory usage tracemalloc
import tracemalloc

def create_large_list():
    large_list = [i for i in range(100000)]

def main():
    tracemalloc.start()
    create_large_list()
    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')
    print("[ Top 10 ]")
    for stat in top_stats[::]:
        print(stat)
    
```

```python
main()
```
