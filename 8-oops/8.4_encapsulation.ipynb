Encapsulation And Abstraction

Encapsulation and abstractio are two  fundamental principles of object-oriented Programming(oop) that help in designing robust, maintainable, and reusable code. Encapsulation invilves buliding data and methods that operate on the data within a single unit, whlie abstractio involves hidling complex implementation details and explosing only the necessar features.


Encapsulation

Encapsulation is the concept of wrapping data(variable) and methods (functions) together as a single unit. it restricts direct access to some of the object's compononents, which is a means of preventing accidental interference and misuse of the data

```python
## Encapsulation with getter and setter Methods
## public, protected, private variables


class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age  

person=person("Ajeeb sagar", 20 )       
print(person.name)
print(person.age)
```

```python
## Encapsulation with getter and setter Methods
## public, protected, private variables


class person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age  

person=person("Ajeeb sagar", 20)       
print(person._person__name)   ## private variable access
print(person._person__age)     ## private variable access
```

```python
p = person.__class__("Ajeeb sagar", 20)
dir(p)
```

```python
### Encapsulation with getter and setter Methods
### public, protected, private variables

class person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age


        ## getter method for name
    def get_name(self):
        return self.__name
        ## getter method for age
    def get_name(self):
        return self.__name

       ##getter method for age
    def get_age(self):
        return self.__age
        ## setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age must be positive")



person = person("Ajeeb sagar", 20)
print(person.get_name())


## Access and modify private variables using getter and setter methods
print(person.get_age())

person.set_age(25)  # Update age using setter
print(person.get_age())  # Check updated age

person.set_age(-5)  # Attempt to set invalid age
```
