SQL AND SQLite

SQL(Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database enginer that is widely used for embedded database systems. In this lesson, we will cover the basic of SQL nad SQLite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3
```

```python
## connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
## crete a table

cursor.execute('''CREATE TABLE IF NOT EXISTS employees
                  (id INTEGER PRIMARY KEY,
                     name TEXT NOT NULL,
                     age INTEGER,
                     department TEXT);''')

## commit the changes
connection.commit()                     
```

```python
cursor.execute('''
             select * from employees;
''')
```

```python
## insert data into the table

cursor.execute('''
             insert into employees (name, age, department)
                values ('Alice', 30, 'HR');
''')

cursor.execute('''
             insert into employees (name, age, department)
                values ('Ajeeb sagar', 21, 'HR');
''')
cursor.execute('''
             insert into employees (name, age, department)
                values ('Ankit', 22, 'HR');
''')
cursor.execute('''
             insert into employees (name, age, department)
                values ('Amit', 34, 'HR');
''')
cursor.execute('''
             insert into employees (name, age, department)
                values ('Alive', 36, 'HR');
''')

## commit the changes
connection.commit()
```

```python
## Query data from the table
cursor.execute('''
             select * from employees;
''')

rows=cursor.fetchall()
## print the queries data

for row in rows:
    print(row)
```

```python
## update data in the table
cursor.execute('''
             update employees   
                set age=31
                where name='Alice';     
''')
## commit the changes
connection.commit()
```

```python
## Query data from the table
cursor.execute('''
             select * from employees;
''')

rows=cursor.fetchall()
## print the queries data

for row in rows:
    print(row)
```

```python
## Delete data from the table
cursor.execute('''
             delete from employees
                where name='Amit';     
''')    
## commit the changes
connection.commit()
```

```python
## Query data from the table
cursor.execute('''
             select * from employees;
''')

rows=cursor.fetchall()
## print the queries data

for row in rows:
    print(row)
```

```python
##working with sales data

## connect to an SQLite database
connection=sqlite3.connect('sales.db')
cursor=connection.cursor()

## create a table for sales data
cursor.execute('''CREATE TABLE IF NOT EXISTS sales
                  (id INTEGER PRIMARY KEY,
                     product_name TEXT NOT NULL,
                     date TEXT NOT NULL,
                     quantity INTEGER,
                     price REAL,
                     region TEXT,
                     sale_date TEXT);''')


## insert data into the sales table
sales=[
    ('Laptop', '2023-10-01', 5, 800.00, 'North', '2023-10-05'),
    ('Smartphone', '2023-10-02', 10, 500.00, 'South', '2023-10-06'),
    ('Tablet', '2023-10-03', 7, 300.00, 'East', '2023-10-07'),
    ('Headphones', '2023-10-04', 15, 100.00, 'West', '2023-10-08'),
    ('Smartwatch', '2023-10-05', 8, 200.00, 'North', '2023-10-09')
]

cursor.executemany('''
             insert into sales (product_name, date, quantity, price, region, sale_date)
                values (?, ?, ?, ?, ?, ?);
''', sales)





## insert sample sales data
cursor.execute("INSERT INTO sales (product_name, date, quantity, price, region, sale_date) VALUES ('Laptop', '2023-10-01', 5, 800.00, 'North', '2023-10-05');")
cursor.execute("INSERT INTO sales (product_name, date, quantity, price, region, sale_date) VALUES ('Smartphone', '2023-10-02', 10, 500.00, 'South', '2023-10-06');")
cursor.execute("INSERT INTO sales (product_name, date, quantity, price, region, sale_date) VALUES ('Tablet', '2023-10-03', 7, 300.00, 'East', '2023-10-07');")
cursor.execute("INSERT INTO sales (product_name, date, quantity, price, region, sale_date) VALUES ('Headphones', '2023-10-04', 15, 100.00, 'West', '2023-10-08');")
cursor.execute("INSERT INTO sales (product_name, date, quantity, price, region, sale_date) VALUES ('Smartwatch', '2023-10-05', 8, 200.00, 'North', '2023-10-09');")

## commit the changes
connection.commit()
```
