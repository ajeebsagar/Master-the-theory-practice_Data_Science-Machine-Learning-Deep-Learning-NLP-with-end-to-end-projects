file handling is a crucial part of anyprogramming language. python provides bulit-in  function and methd to read fromand write to files, both text and binary. this lesson will cover the basics of file handling, including reading and writing text files and binary files.
 

```python
### read a whole file

with open('example.txt', 'r') as file: 
    content=file.read()
    print(content) 
```

```python
## read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())   ### strip() removes the newline character
```

```python
## writing a file(overwriting)

with open('example.txt', 'w') as file:
    file.write("Hello how are you?\n")
    file.write("this is my new line\n")
```

```python
### write a file (appending)

with open('example.txt', 'a') as file:
    file.write("This is an appended line.\n")
    file.write("Appending another line.\n")
```

```python
### writing a list of lines to a file

lines=['first line\n', 'second line\n', 'third line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary files
data =b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
## reading a file into a list
with open('example.txt', 'rb') as file:
    lines = file.read()
    print(lines)
```

```python
## reading a file into a list
with open('example.bin', 'rb') as file:
    lines = file.read()
    print(lines)
```

```python
## read the content froma source text file and write to a destination text file
#copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()
with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)    
```

```python
# # read the text file and count the number of lines, words, characters.
# counting lines, words, and characters in a text file

def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
line_count, word_count, char_count = count_text_file(file_path) 
print(f"Lines: {line_count}, Words: {word_count}, Characters: {char_count}")
       
```

The W+ in python is used to open a file for both reading and writing, if the does not exit, it will be created. if the file exits, its content is truncated (i.e. the file is overwritten)

```python
###  writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write ("Hello, this is a test file.\n")
    file.write ("This file is used for testing file operations.\n")

    ##move the file cursor to the beginning of the file
    file.seek(0)  

    ## read the content of the file
    content = file.read()
    print(content)
```
