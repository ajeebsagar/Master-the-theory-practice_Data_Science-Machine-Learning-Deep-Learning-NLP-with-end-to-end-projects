Pandas-DataFrame And Series


Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. it provides two primary data structures: Series and DataFrames. A Series is a one-dimensional array-like bject, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabuar data structure with labeled axes(rows and columns).

```python
## series
## a  pandas series is a one-dimensional array-like object that can hold various data types, such as integers, floats, strings, etc. It is similar to a column in a spreadsheet or a database table. Each element in a pandas series is assigned an index, which allows for easy access and manipulation of the data.

import pandas as pd
data=[1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
### create a series from a dictionary
data={'q':10,'w':20,'e':30}
series = pd.Series(data)
print(series)
```

```python
data=[1,2,3,4,5]
index=['a','b','c','d','e']
series = pd.Series(data,index=index)
pd.Series(data,index=index)
```

```python
## DataFrame
## create a DataFrame from a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## creaate a data frame from a list of dictionaries
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
