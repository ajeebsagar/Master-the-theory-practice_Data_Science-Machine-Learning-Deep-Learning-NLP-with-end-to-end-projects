Operator Overloading

Operator Overloading allows you to define the behavior of operator(+,-,+,*)  for custom objects, You achieve this by overriding specific magic in your class.

```python
##### common Operator Overloading Magic Mehtods

'''
__add__(self, other):  defines the behavior of the addition operator (+) for your class.
__sub__(self, other):  defines the behavior of the subtraction operator (-) for your class.
__mul__(self, other):  defines the behavior of the multiplication operator (*) for your class.
__truediv__(self, other):  defines the behavior of the division operator (/) for your class.
__floordiv__(self, other):  defines the behavior of the floor division operator (//) for your class.
__mod__(self, other):  defines the behavior of the modulus operator (%) for your class.
__eq__(self, other):  defines the behavior of the equality operator (==) for your class.
__ne__(self, other):  defines the behavior of the inequality operator (!=) for your class.
__lt__(self, other):  defines the behavior of the less than operator (<) for your class.
'''
```

```python
### mathic methods are special methods in Python that allow you to define the behavior of your objects for built-in operations, such as arithmetic operations, comparisons, and more. These methods start and end with double underscores (dunder methods) and can be overridden to customize how your objects behave when used with operators.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __truediv__(self, scalar):
        return Vector(self.x / scalar, self.y / scalar)
    def __eq__(self):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


### create vectors class of the vector
v1 = Vector(2, 3)
v2 = Vector(4, 5)   
print( v1 + v2)
print(v1 - v2)
print(v1 * 2)
print(v1 / 2)
       
```

```python
### overloading operators for complex numbers

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)
    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)
    def __mul__(self, other):
       real_part = self.real * other.real - self.imag * other.imag
       imag_part = self.real * other.imag + self.imag * other.real
       return ComplexNumber(real_part, imag_part)   
    def __truediv__(self, other):
        denominator = other.real ** 2 + other.imag ** 2
        real_part = (self.real * other.real + self.imag * other.imag) / denominator
        imag_part = (self.imag * other.real - self.real * other.imag) / denominator
        return ComplexNumber(real_part, imag_part)
    def __repr__(self):
        return f"{self.real} + {self.imag}i"
    def __eq__(self, other):
        return self.real == other.real and self.imag == other.imag

## create objects of the ComplexNumber class
c1 = ComplexNumber(3, 4)
c2 = ComplexNumber(1, 2)

print(c1 + c2)  # Output: 4 + 6i
print(c1 - c2)  # Output: 2 + 2i
print(c1 * c2)  # Output: -5 + 10i
print(c1 / c2)  # Output: 2.2 + 0.4i
print(c1 == c2) # Output: False
```
