```python
### try, except, else block

try:
    num=int(input("enter a number"))
    result=10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
else:
    print("result is", result)

finally:
    print("this block always executes")
```
