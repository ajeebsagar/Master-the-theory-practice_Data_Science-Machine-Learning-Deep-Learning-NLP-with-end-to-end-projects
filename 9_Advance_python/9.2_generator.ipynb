Generators


Generators are a simpler way to create iterators. they use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.


```python
def square(n):
    for i in range(3):
        yield i **2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

Partical example: reading large files

generators are particularly useful for reading large files because they allow you to  process one line at a time wothout loading the entire file into memory.

```python
### practical : rading large files


def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
file_path = 'large_file.txt'
for line in read_large_file(file_path):
    print(line)
```
