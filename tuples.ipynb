```python
## creating  tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))  # Output: <class 'tuple'>
```

```python
lst=list()
print(type(lst))  # Output: <class 'list'>
tpl=tuple()
print(type(tpl))
```

```python
## immutable nature of tuples
# Tuples are immutable, meaning their elements cannot be changed after creation.

lst = [1, 2, 3,4,5]
print(lst)

lst[1]="sagar"
print(lst)
```

```python
##tuple methods
print(numbers.count(1))  # Output: 1
print(numbers.index(3))  # Output: 1  index is the function  
```

```python
## packing and unpacking tuples
#packing
packed_tuple = (1, "hello", 3.14)
print(packed_tuple)
```

```python
#unpacking
a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with *
numbers=(1,2,3,4,5,6)
first, *middle, last = numbers
print(first)  
print(middle)
print(last) 
```
